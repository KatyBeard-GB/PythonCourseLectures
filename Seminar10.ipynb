**Задача №63**
1. Изобразите отношение households к population с
помощью точечного графика


```python
import pandas as pd
import seaborn as sns

df = pd.read_csv('sample_data/california_housing_train.csv')

sns.scatterplot(data=df, x="households", y="population")
```

2. Визуализировать longitude по отношения к
median_house_value, используя линейный график

```python
sns.relplot(x="longitude", y="median_house_value", kind="line", data=df)
```

3. Представить гистограмму по housing_median_age

```python
sns.histplot(data = df, x = 'housing_median_age')
```

4. Изобразить гистограмму по median_house_value с
оттенком housing_median_age

```python
sns.set()
sns.displot(data=df, x = 'median_house_value', hue="housing_median_age")
```

```python
df.loc[df['median_house_value'] <= 100000, 'value_group'] = 'Econom'
df.loc[(df['median_house_value'] > 100000) & (df['median_house_value'] <= 200000),
'value_group'] = 'Comfort'
df.loc[df['median_house_value'] > 200000, 'value_group'] = 'Premium'
```

```python
df.groupby('value_group')['housing_median_age'].mean().plot(kind='bar')
```

```python
sns.set()
sns.displot(data=df, x = 'housing_median_age', hue="value_group")
```
